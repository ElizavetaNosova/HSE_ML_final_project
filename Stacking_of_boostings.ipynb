```python
import pandas as pd
data = pd.read_csv('toxic-comment-train.csv')
```

```python
data
```

```python
!pip install sklearn
import sklearn
```

```python
from sklearn.model_selection import train_test_split
```

```python
texts = data.comment_text
y = data.toxic
texts_train, texts_test, y_train, y_test = train_test_split(texts, y, random_state=0)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
X_train = tfidf.fit_transform(texts_train)
X_test = tfidf.transform(texts_test)
```

```python
!pip install catboost
```

```python
!pip install lightgbm
```

```python
from lightgbm import LGBMClassifier
```

```python
boosting1_light = LGBMClassifier()
```

```python
boosting1_light.fit(X_train, y_train)
```

```python
sklearn.metrics.roc_auc_score(y_test, boosting1_light.predict_proba(X_test)[:,1])
```

```python
!pip install xgboost
```

```python
from xgboost import XGBClassifier
```

```python
boosting2_xgb = XGBClassifier()
```

```python
boosting2_xgb.fit(X_train, y_train)
```

```python
sklearn.metrics.roc_auc_score(y_test, boosting2_xgb.predict_proba(X_test)[:,1])
```

```python
from catboost import CatBoostClassifier
boosting3_cat = CatBoostClassifier()
boosting3_cat.fit(X_train, y_train)
```

```python
sklearn.metrics.roc_auc_score(y_test, boosting3_cat.predict_proba(X_test)[:,1])
```

```python
!pip install mlxtend
```

```python
from mlxtend.classifier import StackingClassifier
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
sclf = StackingClassifier(classifiers=[boosting1_light, boosting2_xgb, boosting3_cat], 
                          meta_classifier=lr)
```

```python
sclf.fit(X_train, y_train)
```

```python
sklearn.metrics.roc_auc_score(y_test, sclf.predict_proba(X_test)[:,1])
```
