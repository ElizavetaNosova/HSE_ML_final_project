```python
not_toxic = train.query('toxic == 0')
```

```python
groups = defaultdict(list)
```

```python
for index, row in toxic.iterrows():
    toxic_types = []
    for toxic_type in ['severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate']:
        if row[toxic_type]:
            toxic_types.append(toxic_type)
    if toxic_types:
        key = '+'.join(toxic_types)
        groups[key].append(row['comment_text'])
```

```python
for toxic_type in groups:
    print(toxic_type, len(groups[toxic_type]))
```

```python
В чистом виде чаще проявляются obscence и insult
Для дальнейшего анализа возьмём только те комбинации, в которых больше 30 текстов
```

```python
filtered_types = {toxic_type:groups[toxic_type] for toxic_type in groups if len(groups[toxic_type])>30}
```

```python
Посчитаем медианную долю повторяющихся слов
```

```python
def duplicates_ratio(text):
    tokens = [token.strip() for token in text.split()]
    if not tokens:
        return(0)
    else:
        n = 0
        known = set()
        for token in tokens:
            if token in known:
                n += 1
            known.add(token)
        return(n/len(tokens))
```

```python
Для нетоксичных текстов:
```

```python
median([duplicates_ratio(text) for text in not_toxic['comment_text']])
```

```python
Для токсичных текстов:
```

```python
for toxic_type in filtered_types:
    print(toxic_type, median([duplicates_ratio(text) for text in filtered_types[toxic_type]]))
```

```python
Я ожидала, что в нецензурных текстах будет больше повторов, но была не права. Даже наоборот, в нетоксичных текстов больше повторов слов, чем в большинстве типов токсичных.
Почему-то этот показатель ощутимо выше в текстах, совмещающих несколько типов токсичности
```

Средняя длина слова

```python
def mean_word_len(text):
    tokens = [token.strip() for token in text.split()]
    if not tokens:
        return 0
    else:
        return len(''.join(tokens))/len(tokens)
```

```python
median([mean_word_len(text) for text in not_toxic['comment_text']])
```

```python
for toxic_type in filtered_types:
    print(toxic_type, median([mean_word_len(text) for text in filtered_types[toxic_type]]))
```

```python
Для всех категорий средняя длина слова меньше, чем для нетоксичных текстов, но совсем чуть-чуть
```

```python
Средняя длина предложения
```

```python
from nltk import sent_tokenize
```

```python
def mean_sent_len(text):
    sents = sent_tokenize(text, 'english')
    sent_lens = [len(sent.split()) for sent in sents]
    return mean(sent_lens)
```

```python
median([mean_sent_len(text) for text in not_toxic['comment_text']])
```

```python
for toxic_type in filtered_types:
    print(toxic_type, median([mean_sent_len(text) for text in filtered_types[toxic_type]]))
```
